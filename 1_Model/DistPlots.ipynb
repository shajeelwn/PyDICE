```python
with open('ecs_dist_v4.json') as f:
    d=json.load(f)

t = np.linspace(-5,15, 2000)
plt.figure(figsize=(10,5))
plt.plot(t,stats.norm.pdf(t, d['norm'][0],d['norm'][1]), color = sns.color_palette("inferno", 15)[-3], label = 'Normal')
plt.plot(t,stats.lognorm.pdf(t, d['lognorm'][0],d['lognorm'][1],d['lognorm'][2]), color =sns.color_palette("viridis", 10)[-4], label = 'Log-Normal')
plt.plot(t,stats.cauchy.pdf(t, d['cauchy'][0],d['cauchy'][1]), color = sns.cubehelix_palette(10)[5], label = 'Cauchy')
plt.legend()
```

```python
with open('ecs_dist_v4.json') as f:
    d=json.load(f)

t = np.linspace(-5,15, 2000)
plt.figure(figsize=(10,5))
plt.plot(t,stats.norm.pdf(t, d['norm'][0],d['norm'][1]), color = sns.color_palette("inferno", 15)[-3], label = 'Normal')
plt.plot(t,stats.lognorm.pdf(t, d['lognorm'][0],d['lognorm'][1],d['lognorm'][2]), color =sns.color_palette("viridis", 10)[-4], label = 'Log-Normal')
plt.plot(t,stats.cauchy.pdf(t, d['cauchy'][0],d['cauchy'][1]), color = sns.cubehelix_palette(10)[5], label = 'Cauchy')
plt.legend()
plt.xlim(6,15)
plt.ylim(0,0.1)
```
